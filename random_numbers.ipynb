```python
import numpy as np
import matplotlib.pyplot as plt

n = 10000000

# Generate 100 random numbers uniformly distributed between 0 and 1
random_numbers = 2*np.random.random(n)-1

# Plot histogram
plt.hist(random_numbers, bins=10, edgecolor='black')
plt.title(f'Distribution of {n} Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000000

# Generate n random numbers with a normal distribution
normal_random_numbers = np.random.normal(loc=0.0, scale=1.0, size=n)

# Plot histogram
plt.hist(normal_random_numbers, bins=1000, edgecolor='black')
plt.title(f'Distribution of {n} Normally Distributed Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
n = 10000000

# Generate n random numbers with an exponential distribution
exponential_random_numbers = np.random.exponential(scale=1.0, size=n)

# Plot histogram
plt.hist(exponential_random_numbers, bins=100, edgecolor='black', log=True)
plt.title(f'Distribution of {n} Exponentially Distributed Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency (log scale)')
plt.show()

# Plot histogram without log scale
plt.hist(exponential_random_numbers, bins=100, edgecolor='black')
plt.title(f'Distribution of {n} Exponentially Distributed Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
n = 10000000
lambda_param = 5

# Generate n random numbers with a Poisson distribution
poisson_random_numbers = np.random.poisson(lam=lambda_param, size=n)

# Plot histogram
plt.hist(poisson_random_numbers, bins=range(0, max(poisson_random_numbers)+1), edgecolor='black')
plt.title(f'Distribution of {n} Poisson Distributed Random Numbers (λ={lambda_param})')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
