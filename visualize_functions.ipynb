# Visualize Functions

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display, clear_output

# List of functions and their labels
functions = [
    (lambda x: np.exp(x), 'e^x'),
    (lambda x: np.exp(-x), 'e^(-x)'),
    (lambda x: -np.exp(x), '-e^x'),
    (lambda x: -np.exp(-x), '-e^(-x)')
]

# Representative input values
input_values = np.array([-2, -1, 0, 1, 2])

# Generate x values for plotting
x_values = np.linspace(-2, 2, 400)

# Plot functions
plt.figure(figsize=(10, 6))
for func, label in functions:
    y_values = func(x_values)
    plt.plot(x_values, y_values, label=label)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Visualization of Functions')
plt.legend()
plt.ylim(-3, 3)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()

# Update table
data = {'Input (x)': input_values}
for func, label in functions:
    data[label] = np.round(func(input_values), 2)
df = pd.DataFrame(data)
display(df)
```
